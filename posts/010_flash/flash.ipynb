Gemini just released a new model that will handle image editing that preserves previous qualities in the original image. I am curious to test out the limits of this model to see how its applications could be most useful. In past image generation tools, any attempts to replicate or edit an image would produce an entirely new image with potentially similar qualities. Flash is supposed to now be able to take an image input and add on top, or edit qualities within the image. Let's test it out!

First, I will present an image, and ask for a single modification, in this case the addition of a party hat to a cute puppy dog.

This worked just about as perfectly as I could imagine it! Let's move on to more advanced changes. 

Now, I want to completely change the subject of the image. Because of the proportions and direction of the image, I will keep it a dog, but I'll change the breed. 

That worked great as well!

Finally, I want to complicate things further by asking 2.0 Flash to change two huge parts of the image at once, inherently reinventing the entire photo. This is where things went wrong. Instead of communicating its limitations or offering a viable alterantive to my requests, it encountered an error and was unable to complete my request.

## Discussion
I found this to be a very interesting exploration of this new image generation editing technology. I am curious to test these same features in the feature if more advanced models get released. It seems to handle single basic task edits very well and more complicated or holisitic edits not as well.
