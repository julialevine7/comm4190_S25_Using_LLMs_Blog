## gpt-4o Tasks Beta
OpenAI has recently come out with a new model that is supposed to be able to schedule tasks, such as reminding you of important events, following up on a question, or running a code debugger. This new and versatile tool seems to have promising potential! However, being the LLM experimenter I am, I am interested to see its limits. 

My plan is to test its timing, computation, and generation capabilities. First, I will ask it to complete a task in one second, to which I expect some sort of bypass to the task generation system where gpt-4o will just provide the answer immediately. Then, I will increase the time limit to one minute and ask it to generate an image and conduct a simple calculation.
